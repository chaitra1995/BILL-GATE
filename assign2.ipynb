```python
1.	what are the characteristics of the tuple ?is tuple immutable?
a. Tuples are ordered,indexed collection of data.similar to string indices, the  first value in tuple  will have the index[0],the second value[1] and so on.
b. Tuples  can store  duplicate values.
c. once the  data is assigned the value cannot be changed .
d. tuples allow you to store several data items  in  one variable. you can choose  to store only one kind of data in a tuple  or mix it up as needed.
Yes it is immutable.

2.	 what are the two tuple methods in python ? give an example of each method .give  a reason why tuple have only  two built in methods as compared to lists.
               The two tuple methods are :  a. index()   b. count()
        my_tuple=(‘a’,’p’,’p’,’l’,’e’,)
    print(my_tuple.count(‘p’))
    print(my_tuple.index(‘a’))
in python tuple is immutable . you cannot change the item once is assigned .there are only  two tuple() methods index() &count()that tuple object can call.

3.	which collection datatypes in python  do not allow  duplicate  items? write  a  code  using  asset to remove  duplicates from the  given list.
List=[1,1,1,2,1,3,1,4,2,1,2,2,2,2,3,2,4,3,1,3,2,3,3,3,4,3,4,4,1,4,2,4,3,4,4]

Set {}  is the collection of datatype  which is unindexed, unordered. No duplicates allowed
a={1,1,1,2,1,3,1,4,2,1,2,2,2,2,3,2,4,3,1,3,2,3,3,3,4,3,4,4,1,4,2,4,3,4,4}
print(a)

4.	Explain the difference between the union and update() method for set .give an example of each method.
Union of the two sets A and  B  include all the elements of  set A and B. we use | and union()  method to perform for set.
a={0,2,4}
b={9,5,6,8}
print('union of the set |:',a|b)
and 
print('union of the set ():',a.union(b))

update() method is used to update the  with items  other collection type.
companies={'bangalore','mangalore','chennai'}
tech_companies={'hubli','dharwad','gulburga'}
companies.update(tech_companies)
print(companies)
{'mangalore', 'hubli', 'bangalore', 'gulburga', 'chennai', 'dharwad'}

5.what is a dictionary? give an example .Also  state whether a dictionary is ordered or unordered.
The dictionary is the unordered collection that contains key: value pairs separated by commas inside curly brackets. dictionary  are  optimized to retrieve values when the keys is known
               Capitals={“USA”:”Washington D.C”,”France”:”paris”,”India”:”New Delhi”
       Capital is the dictionary object which contains key –value  pairs inside{}.The left side of : is a key, and the right side  is a value .the should be unique and an immutable object.

6.can we create nested dictionary ? if so, please give  me an example by creating a  simple  one-level  nested  dictionary.  
Yes. A  nested python  dictionary is a dictionary  that has  many  dictionaries . we can create a dictionary 
With that contain a number of dictionaries. With that ,we can also nest the number of dictionaries already existing as dictionaries. Let us first create  a python dictionary and fetch all the values. Here we have included 4 key-value pairs in the dictionary and displayed them. Product, model, units, and available are keys of the dictionary. Except the units key ,all  are having string values-
Creating a dictionary
Myprod={
“product “:”mobile”,
  “model”:”XUT”,
   “units”:120,
“Available”:”yes”
}
Display the dictionary 
Print(Myprod)


7.Using  setdefault () method ,create key named topics in the given dictionary and also add the value of  the key as the list [‘python’,’machine learning’, ‘deep learning’]
Dict1={‘language’:’python’,’course’:’data science’}

Python  setdefault()method  is used to set default value to key .it returns value, if  the key is present .otherwise it returns the default value the key is none.
dict1={'language':'python','course':'data science'}
p=dict1.setdefault('python','machine learning')
print(dict1)
{'language': 'python', 'course': 'data science', 'python': 'machine learning'}

8.what are the three view objects in dictionaries?use the three built-in  method in python to display 
These three view  objects for the given dictionary.
Dict1={‘sport’:’cricket’,’teams’:[‘india’,’austraila’,’england’,’south africa’,’sri lanka’,’new zealand’]}

The three view objects in dictionaries are key, values, items.
a.	Key(): The view object key that represent all the  keys in dictionary .you can retrieve the contents of this object using the keys() method . if you  print the view  object of the keys , it displays  new list of all the keys in the dictionaries in order of insertion .once we retrieve all the keys  using key()method  into variable ,if we update the dictionary the changes wil be reflexed in the variable 
Dict1={'sport':'cricket'}
keys= Dict1.keys()
Dict1['teams']='india','austrila','south africa','srilanka','england'
print("listing updated dictionary:",Dict1)
print("keys after updated:",keys)
listing updated dictionary: {'sport': 'cricket', 'teams': ('india', 'austrila', 'south africa', 'srilanka', 'england')}
keys after updated: dict_keys(['sport', 'teams'])

b.	Values():  is the built in method of python  programming language  which returns a view  object. That view  object contains the value  of dictionary  as a list.
Dict1={'sport':'cricket'}
values= Dict1.values()
Dict1['teams']='india','austrila','south africa','srilanka','england'
print("listing updated dictionary:",Dict1)
print("values after updated:",values)
listing updated dictionary: {'sport': 'cricket', 'teams': ('india', 'austrila', 'south africa', 'srilanka', 'england')}
values after updated: dict_values(['cricket', ('india', 'austrila', 'south africa', 'srilanka', 'england')])
      
c.	Items():method returns  a new  view of dictionary’s  items((key,value),pairs)as a list  with all dictionary  key with values.
Dict1={'sport':'cricket'}
items= Dict1.items()
Dict1['teams']='india','austrila','south africa','srilanka','england'
print("listing updated dictionary:",Dict1)
print("values after updated:",items)
listing updated dictionary: {'sport': 'cricket', 'teams': ('india', 'austrila', 'south africa', 'srilanka', 'england')}
values after updated: dict_items([('sport', 'cricket'), ('teams', ('india', 'austrila', 'south africa', 'srilanka', 'england'))])



   

```
